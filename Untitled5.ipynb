```python
imgs = []
for i in range(1, 200001):
  name = str(i)+".jpg"
  img = Image.open("/train_pics_ew_200000/"+name)
  imgs.append(img)
```

```python
labels = pd.read_csv("/train_new.csv")
```

```python
class Baidu_dataset(torch.utils.data.Dataset):
  def __init__(self, transforms, labels, train): 
    self.transforms = transforms 
    self.train = train
    self.labels = pd.read_csv(labels)
  def __getitem__(self, index): 
    if self.train == True: 
      img = imgs[index]
      img = self.transforms(img)
      return img, self.labels["train"][index]
    else:
      name = str(index+1)+".jpg"
      img = Image.open("/content/drive/My Drive/test_pics/"+name)
      img = self.transforms(img)
      return img, self.labels["test"][index]
  def __len__(self): 
    if self.train == True:
      return 200000
    else:
      return 10000
```

```python
my_transform = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Resize(299),
    torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
train_dataset = Baidu_dataset(my_transform, "/train_new.csv", train=True)
```

```python
model = torch.hub.load('pytorch/vision:v0.10.0', 'resnet152')
```

```python
train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
device = torch.device("cuda")
```

```python
model.fc = torch.nn.Linear(in_features=2048, out_features=1, bias=True)
```

```python
model = model.to(device)
```

```python
def loss_function(output, label): 
  output = torch.arctan(output)*(180/torch.pi)
  loss1 = torch.abs(output-label)
  loss = torch.Tensor(loss1).mean().view(1)
  loss.requires_grad_(True)
  return loss
```

```python
from tqdm import tqdm
import time
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer=optimizer, step_size=30, gamma=0.1)
epochs = 90
model.train()
for epoch in tqdm(range(epochs)): 
  average_loss = 0
  for idx, (img, label) in enumerate(train_dataloader): 
    img, label = img.to(device), label.to(device)
    model.zero_grad()
    output = model(img).view(32,1)
    loss = loss_function(output, label.unsqueeze(1))
    loss.backward()
    optimizer.step()
    average_loss+=torch.sum(loss).item()

  lr_scheduler.step()
  print("Epoch:{}\t average_loss:{}".format(epoch+1, average_loss/idx))

torch.save(model, "/baidu_new_tan.pkl")
```
